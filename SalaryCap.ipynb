# Contract Amount

```python
contracts_ten = contracts[(contracts['team']=='Titans') & (contracts['is_active']==True)]
contracts_ten_position = contracts_ten.groupby(by=['position','team'],as_index=False).agg({'apy':'sum'})
contracts_nfl = contracts[(contracts['is_active']==True)]
contracts_nfl = contracts_nfl.groupby(by=['position','team'], as_index=False).agg({'apy':'sum'})
contracts_nfl = contracts_nfl.groupby(by=['position'], as_index=False).agg({'apy':'mean'})
contracts_ten_position = contracts_ten_position.merge(team_colors,how='left',left_on=['team'],right_on=['team_nick'])
contract_all = contracts_ten_position.merge(contracts_nfl,on='position',suffixes=['_ten','_nfl'],how='outer')
```

```python
fig = plt.figure(figsize=(10,6),facecolor='#ffff')

X_axis = np.arange(len(contract_all)) 

plt.bar(x=X_axis-0.35,height=contract_all['apy_ten'],label='TEN',color=contract_all[~contract_all['team_color2'].isna()]['team_color2'],width=0.35)
plt.bar(x=X_axis,height=contract_all['apy_nfl'],label='NFL',color='lightgrey',width=0.35)
plt.xticks(X_axis, contract_all['position']) 

# plt.title('How are the Titans using their cap space?',fontsize=20)
plt.ylabel('APY (millions)',fontsize=20)
plt.xlabel('Position',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.legend()

# plt.show()

plt.suptitle('How are the Titans using their cap space?',fontsize=20)
plt.title('2024 Season | Data: nfl_data_py | Chart: Matt Flaherty (@flahermi)',y=1,x=0.5,fontsize=10)
plt.savefig(
    './visuals/capSpacePosition.png',
#     transparent=False,
    bbox_inches='tight',
#     facecolor=fig.get_facecolor('#ffff')
)
```

Having 2 starting O-linemen and a starting QB on rookie contracts has allowed the Titans to spend more than league average on CB, DL, and WR.

# Snap % vs Cap %

```python
snap_count.head()
```

```python
snap_count_ten = snap_count[snap_count['team']=='TEN']
snap_count_ten = snap_count_ten.groupby(by='player',as_index=False).agg({'offense_pct':'mean','defense_pct':'mean'})
snap_cap = snap_count_ten.merge(contracts_ten, how='left', on=['player'])
snap_cap['snap_pct'] = snap_cap['offense_pct'] + snap_cap['defense_pct']
```

```python
fig = plt.figure(figsize=(10,6),facecolor='#ffff')

# vals = ax.get_yticks()
# plt.gca().set_yticklabels(['{:,.2%}'.format(x) for x in snap_cap['apy_cap_pct']])
# plt.gca().set_xticklabels(['{:,.2%}'.format(x) for x in snap_cap['snap_pct']])

plt.scatter(x=snap_cap['snap_pct'],y=snap_cap['apy_cap_pct'])
for i in range(snap_cap.shape[0]):
    if (snap_cap['apy_cap_pct'][i]>0.04) | (snap_cap['snap_pct'][i]>0.8):
        plt.annotate(snap_cap['player'][i],xy=(snap_cap['snap_pct'][i],snap_cap['apy_cap_pct'][i]),fontsize=15)
        
# plt.title('Does Contract Value Match Snap Percent?',fontsize=20)
plt.ylabel('Cap %',fontsize=20)
plt.xlabel('Snap %',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.suptitle('Does Contract Value Match Snap Percent?',fontsize=20)
plt.title('2024 Season | Data: nfl_data_py | Chart: Matt Flaherty (@flahermi)',y=1,x=0.5,fontsize=10)
plt.savefig(
    './visuals/capSnapPct.png',
#     transparent=False,
    bbox_inches='tight',
#     facecolor=fig.get_facecolor('#ffff')
)
```
