```python
import pandas as pd
import numpy as np
import nfl_data_py as nfl

import matplotlib.pyplot as plt 

pd.set_option('display.max_columns', 500)
```

```python
contracts = pd.read_parquet('C:/Users/19012/Documents/TitansOffseason2025/data/historical_contracts.parquet', engine='fastparquet')
contracts.head()
```

```python
rosters = pd.read_parquet('C:/Users/19012/Documents/TitansOffseason2025/data/roster_2024.parquet', engine='fastparquet')
rosters.head()
```

```python
team_colors = nfl.import_team_desc()
```

```python
contracts.merge(rosters[(rosters['team']=='TEN') & (rosters['status']=='ACT')], how='inner', on=['gsis_id'])
```

```python
contracts_nfl = contracts[(contracts['is_active']==True)]
contracts_nfl = contracts_nfl.groupby(by=['position','team'], as_index=False).agg({'apy':'sum'})
contracts_nfl.groupby(by=['position'], as_index=False).agg({'apy':'mean'})
```

```python
contracts_ten = contracts[(contracts['team']=='Titans') & (contracts['is_active']==True)]
contracts_ten = contracts_ten.groupby(by=['position','team'],as_index=False).agg({'apy':'sum'})
contracts_nfl = contracts[(contracts['is_active']==True)]
contracts_nfl = contracts_nfl.groupby(by=['position','team'], as_index=False).agg({'apy':'sum'})
contracts_nfl = contracts_nfl.groupby(by=['position'], as_index=False).agg({'apy':'mean'})
contracts_ten = contracts_ten.merge(team_colors,how='left',left_on=['team'],right_on=['team_nick'])
contract_all = contracts_ten.merge(contracts_nfl,on='position',suffixes=['_ten','_nfl'],how='outer')
```

```python
fig = plt.figure(figsize=(10,6),facecolor='#ffff')

X_axis = np.arange(len(contract_all)) 

plt.bar(x=X_axis-0.35,height=contract_all['apy_ten'],label='TEN',color=contract_all[~contract_all['team_color2'].isna()]['team_color2'],width=0.35)
plt.bar(x=X_axis,height=contract_all['apy_nfl'],label='NFL',color='lightgrey',width=0.35)
plt.xticks(X_axis, contract_all['position']) 

plt.title('How are the Titans using their cap space?',fontsize=20)
plt.ylabel('APY (millions)',fontsize=20)
plt.xlabel('Position',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.legend()

# plt.show()

plt.savefig(
    './visuals/capSpacePosition.png',
#     transparent=False,
    bbox_inches='tight',
#     facecolor=fig.get_facecolor('#ffff')
)
```
