# Contract Amount

```python
contracts_ten = contracts[(contracts['team']=='Titans') & (contracts['is_active']==True)]
contracts_ten_position = contracts_ten.groupby(by=['position','team'],as_index=False).agg({'apy':'sum'})
contracts_nfl = contracts[(contracts['is_active']==True)]
contracts_nfl = contracts_nfl.groupby(by=['position','team'], as_index=False).agg({'apy':'sum'})
contracts_nfl = contracts_nfl.groupby(by=['position'], as_index=False).agg({'apy':'mean'})
contracts_ten_position = contracts_ten_position.merge(team_colors,how='left',left_on=['team'],right_on=['team_nick'])
contract_all = contracts_ten_position.merge(contracts_nfl,on='position',suffixes=['_ten','_nfl'],how='outer')
```

```python
fig = plt.figure(figsize=(10,6),facecolor='#ffff')

X_axis = np.arange(len(contract_all)) 

plt.bar(x=X_axis-0.35,height=contract_all['apy_ten'],label='TEN',color=contract_all[~contract_all['team_color2'].isna()]['team_color2'],width=0.35)
plt.bar(x=X_axis,height=contract_all['apy_nfl'],label='NFL',color='lightgrey',width=0.35)
plt.xticks(X_axis, contract_all['position']) 

# plt.title('How are the Titans using their cap space?',fontsize=20)
plt.ylabel('APY (millions)',fontsize=20)
plt.xlabel('Position',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.legend()

# plt.show()

plt.suptitle('How are the Titans using their cap space?',fontsize=20)
plt.title('2024 Season | Data: nfl_data_py | Chart: Matt Flaherty (@flahermi)',y=1,x=0.5,fontsize=10)
plt.savefig(
    './visuals/capSpacePosition.png',
#     transparent=False,
    bbox_inches='tight',
#     facecolor=fig.get_facecolor('#ffff')
)
```

Having 2 starting O-linemen and a starting QB on rookie contracts has allowed the Titans to spend more than league average on CB, DL, and WR.

```python
wr_epa_breakdown = nfl_pbp[(nfl_pbp['play_type']=='pass') & (nfl_pbp['sack']!=1) & (nfl_pbp['down'].notna())].copy()
wr_epa_breakdown['pass_length_new'] = np.where(wr_epa_breakdown['air_yards']>20,'long','short')
wr_epa_breakdown = wr_epa_breakdown.\
                        groupby(by=['receiver_player_id','receiver_player_name','pass_length_new'],as_index=False).\
                        agg({'epa':['mean','count']})
wr_epa_breakdown.columns = list(map('_'.join,wr_epa_breakdown.columns))
wr_epa_breakdown = wr_epa_breakdown.merge(rosters[['gsis_id','position']],how='left',left_on='receiver_player_id_',right_on=['gsis_id'])
wr_epa_breakdown = wr_epa_breakdown[wr_epa_breakdown['position']=='WR']
# wr_epa_breakdown.head()
wr_epa_breakdown_short = wr_epa_breakdown[wr_epa_breakdown['pass_length_new_']=='short']
wr_epa_breakdown_short = wr_epa_breakdown_short[wr_epa_breakdown_short['epa_count']>=40].reset_index(drop=True)
wr_epa_breakdown_short[wr_epa_breakdown_short['receiver_player_name_']=='N.Westbrook-Ikhine']
```

```python
wr_epa_breakdown_long = wr_epa_breakdown[wr_epa_breakdown['pass_length_new_']=='long']
wr_epa_breakdown_long = wr_epa_breakdown_long[wr_epa_breakdown_long['epa_count']>=7].reset_index(drop=True)
wr_epa_breakdown_long[wr_epa_breakdown_long['receiver_player_name_']=='N.Westbrook-Ikhine']
```

```python
wr_epa_breakdown = nfl_pbp[(nfl_pbp['play_type']=='pass') & (nfl_pbp['sack']!=1) & (nfl_pbp['down'].notna()) & (nfl_pbp['touchdown']==1)].copy()
wr_epa_breakdown = wr_epa_breakdown.\
                        groupby(by=['receiver_player_id','receiver_player_name'],as_index=False).\
                        agg({'epa':['mean','count']})
wr_epa_breakdown.columns = list(map('_'.join,wr_epa_breakdown.columns))
wr_epa_breakdown = wr_epa_breakdown.merge(rosters[['gsis_id','position']],how='left',left_on='receiver_player_id_',right_on=['gsis_id'])
wr_epa_breakdown = wr_epa_breakdown[wr_epa_breakdown['position']=='WR']
wr_epa_breakdown = wr_epa_breakdown[wr_epa_breakdown['epa_count']>=3].reset_index(drop=True)
wr_epa_breakdown[wr_epa_breakdown['receiver_player_name_']=='N.Westbrook-Ikhine']
```

I think the Titans keep NWI because he was able to be a top 50 pass cather (EPA/target WRs with 60+ targets) following right behind Ridley and Hopkins. The 45th-55th highest contracts APY to WRs in 2024 was \$4-\$6 million; therefore, I think the Titans can get away with paying him near this amount because they signed him to a 1-year contract for 2 million in 2024. This rewards his performance while not overpaying for him. It also sounds like the Titans will not be big spenders in free agency this year.

NWI's EPA/target ranked about 30th in both short and long passes which shows that he can succeed in both ranges. If NWI had been ranked high in EPA/target for long passes and low in EPA/target in short passes, you'd have to question whether his production was sustainable. He also ranked in roughly the same spot for EPA/touchdown.

# Snap % vs Cap %

```python
snap_count.head()
```

```python
snap_count = snap_count.merge(rosters,how='left',left_on='pfr_player_id',right_on='pfr_id',suffixes=['','_roster'])
snap_count_ten = snap_count[snap_count['team']=='TEN']
snap_count_ten = snap_count_ten.groupby(by=['player','pfr_player_id'],as_index=False).agg({'offense_pct':'mean','defense_pct':'mean'})
snap_cap = snap_count_ten.merge(contracts_ten, how='left', on=['player'])
snap_cap['snap_pct'] = snap_cap['offense_pct'] + snap_cap['defense_pct']
```

```python
fig = plt.figure(figsize=(10,6),facecolor='#ffff')

# vals = ax.get_yticks()
# plt.gca().set_yticklabels(['{:,.2%}'.format(x) for x in snap_cap['apy_cap_pct']])
# plt.gca().set_xticklabels(['{:,.2%}'.format(x) for x in snap_cap['snap_pct']])

plt.scatter(x=snap_cap['snap_pct'],y=snap_cap['apy_cap_pct'])
for i in range(snap_cap.shape[0]):
    if (snap_cap['apy_cap_pct'][i]>0.04) | (snap_cap['snap_pct'][i]>0.8):
        plt.annotate(snap_cap['player'][i],xy=(snap_cap['snap_pct'][i],snap_cap['apy_cap_pct'][i]),fontsize=15)
        
# plt.title('Does Contract Value Match Snap Percent?',fontsize=20)
plt.ylabel('Cap %',fontsize=20)
plt.xlabel('Snap %',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.suptitle('Does Contract Value Match Snap Percent?',fontsize=20)
plt.title('2024 Season | Data: nfl_data_py | Chart: Matt Flaherty (@flahermi)',y=1,x=0.5,fontsize=10)
plt.savefig(
    './visuals/capSnapPct.png',
#     transparent=False,
    bbox_inches='tight',
#     facecolor=fig.get_facecolor('#ffff')
)
```
